# _Transformation argent produits_

```python
from pyspark.sql.types import *
from pyspark.sql.functions import *

dfProducts=spark.read.parquet("Files/Bronze/products.parquet")
dfProducts=dfProducts.select("id","code","product_name","image_url","product_quantity","product_quantity_unit","brands",
                            "nutriscore_grade","ecoscore_grade","nova_group","created","updated")
```

```python
#Vérification du type des colonnes
dfProducts.printSchema()
```

```python
#Conversion de product_quantity et nova_group en entier
dfProducts=dfProducts.withColumn("product_quantity",col("product_quantity").cast("int")).withColumn("nova_group",dfProducts.nova_group.cast(IntegerType()))
#Conversion de created et updated en type date
dfProducts=dfProducts.withColumn("created",dfProducts.created.cast(DateType())).withColumn("updated",dfProducts.updated.cast(DateType()))
dfProducts=dfProducts.withColumn("id",dfProducts.id.cast(StringType()))
```

```python
dfProducts.printSchema()
```

```python
dfProducts.write.format("parquet").mode("overwrite").save("Files/Argent/Transformproducts.parquet")
```

# _Transformation argent Prix_

```python
dfPrices=spark.read.parquet("Files/Bronze/Prices.parquet")
dfPrices=dfPrices.select('created',
 'currency',
 'date',
 'id',
 'location_id',
 'price',
 'price_is_discounted',
 'price_per',
 'price_without_discount',
 'product_id',
 'receipt_quantity',
 'updated')
```

```python
#Conversion des colonnes au format approprié
dfPrices=dfPrices.withColumn("date",dfPrices.date.cast(DateType())).withColumn("created",dfPrices.created.cast(DateType()))
dfPrices=dfPrices.withColumn("updated",dfPrices.updated.cast(DateType()))
dfPrices=dfPrices.withColumn("id",dfPrices.id.cast(StringType())).withColumn("location_id",dfPrices.location_id.cast(StringType())).withColumn("product_id",dfPrices.product_id.cast(StringType()))
```

```python
dfPrices.printSchema()
```

```python
dfPrices.write.format("parquet").mode("overwrite").save("Files/Argent/TransformPrices.parquet")
```

# _Transformations argent Locations_

```python
dfLocations=spark.read.parquet("Files/Bronze/Locations.parquet")
dfLocations=dfLocations.select('created',
 'id',
 'osm_address_city',
 'osm_address_country',
 'osm_address_country_code',
 'osm_address_postcode',
 'osm_brand',
 'osm_display_name',
 'osm_id',
 'osm_lat',
 'osm_lon',
 'osm_name',
 'osm_tag_key',
 'osm_tag_value',
 'type',
 'updated',
 'website_url')
```

```python
for column in dfLocations.columns:
    if column in ["date","updated","created"]:
        dfLocations = dfLocations.withColumn(column,dfLocations[column].cast(DateType()))

dfLocations.printSchema()

dfLocations=dfLocations.withColumn("id",dfLocations.id.cast(StringType()))
```

```python
dfLocations.write.format("parquet").mode("overwrite").save("Files/Argent/TransformLocation.parquet")
```
