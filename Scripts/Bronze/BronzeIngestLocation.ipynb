# _IMPORTATION DES LOCALISATIONS_

```python
import requests
import aiohttp
import asyncio
import pandas as pd
from tqdm.asyncio import tqdm

size=100

locationPagesUrl=f"https://prices.openfoodfacts.org/api/v1/locations?size={size}"

responseLocationPages=requests.get(locationPagesUrl)
locationPagesJson=responseLocationPages.json()
lastPageLocation=locationPagesJson["pages"]
print(f"le nombre de page total de localisation pour une taille de {size} est:{lastPageLocation}")
```

```python
async def getLocations(url):
    locationList=[]
    async with aiohttp.ClientSession() as session:
        for page in tqdm(range(1,lastPageLocation+1)):
            responseLocations= await session.get(url.format(size,page))
            locationsJson= await responseLocations.json()
            locationList.extend(locationsJson["items"])
    return locationList
```

```python
locationUrl="https://prices.openfoodfacts.org/api/v1/locations?size={}&page={}"
locationList= await getLocations(locationUrl)
```

```python
#dfLocation=pd.json_normalize(locationList)
dfLocationSpark=spark.createDataFrame(locationList)
dfLocationSpark.write.format("parquet").mode("overwrite").save("Files/Bronze/Locations.parquet")
#display(dfLocationSpark)
```
